The objective of a movie recommendation system is to predict user preferences for movies and suggest movies that they are likely to enjoy.

This is done by analyzing past user behavior, such as ratings or viewing history, and identifying patterns to recommend relevant movies.

So our motive is to tell what movie are most similar to the user's movie choice

```python
mov_features = mov[['Movie_ID',
       'Movie_Budget', 'Movie_Popularity', 'Movie_Release_Date',
       'Movie_Revenue', 'Movie_Runtime']].fillna('')
```

```python
mov_features.shape
mov_features
```

```python
# Convert numerical columns to strings
mov_features['Movie_Budget'] = mov_features['Movie_Budget'].astype(str)
mov_features['Movie_Popularity'] = mov_features['Movie_Popularity'].astype(str)
mov_features['Movie_Revenue'] = mov_features['Movie_Revenue'].astype(str)
mov_features['Movie_Runtime'] = mov_features['Movie_Runtime'].astype(str)

# Combine the features
x = mov_features['Movie_Budget']+ ' ' + mov_features['Movie_Popularity']+ ' '+ mov_features['Movie_Release_Date']+ ' ' +mov_features['Movie_Revenue']+ ' '+mov_features['Movie_Runtime']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
hi = TfidfVectorizer()
```

```python
x = hi.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(x)
```

```python
sim
```

```python
Favourite_Movie_Name = input('Enter your favourite movie name : ')
```

```python
All_movie = mov['Movie_Title'].tolist()
```

```python
import difflib
```

```python
movierec = difflib.get_close_matches(Favourite_Movie_Name, All_movie)
```

```python
print(movierec)
```

```python
close = movierec[0]
```

```python
sorted_movie = sorted(list(enumerate(sim[0])), reverse = True, key = lambda x:x[1])
```

```python
print("top 20 movie name")
```

```python
i = 1
for movie in sorted_movie:
  index = movie[0]
  title_from_index = mov[mov.index==index]['Movie_Title'].values[0]
  if(i<20):
    print(i, '.',title_from_index)
    i = i+1
  else:
    break
```
